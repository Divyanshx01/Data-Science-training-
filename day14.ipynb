```python
# Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex' , hue = 'alive', palette = sns.color_palette('magma',2))  # hue: further divide: based on categorical(object)
plt.title('Titanic Survival Analysis')
plt.show()
```

```python
sns.barplot(data = df_titanic, x = 'class', y = 'fare', hue = 'alive',errorbar= ('ci',0),palette = sns.color_palette('magma',2))
plt.grid(color = 'r',alpha = 0.1)
plt.title('Titanic Survial')
plt.show()
```

```python
# sns.boxplot
plt.figure(figsize = (10,5))
sns.boxplot(data = df_titanic, y='class', x   = 'fare' , hue = 'sex',orient = 'h')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.violinplot(data = df_titanic, x='class', y   = 'fare' , hue = 'sex',orient = 'v')
plt.show()
```
