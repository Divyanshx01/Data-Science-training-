```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print('Done')
```

```python
data = {'Study_hrs':range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}

df = pd.DataFrame(data)
df
```

```python
# ML model build?: x, y: +ve, -ve
# Scatter Plot
plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker = 'X', color = 'c')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

```python
# Import Module
from sklearn.linear_model import LinearRegression

# Object Model Creation
model = LinearRegression()


# X, y
X = df[['Study_hrs']]
y = df['Marks Obtained']


# Training/Learning
model.fit(X,y)
```

```python
m = model.coef_
c = model.intercept_
print(m,c)
```

```python

std_hrs = [[5.5]]

marks = m * std_hrs + c
# y = mx + c
# "y =  + x"
# y = wx +  c

print('Predicted Marks: ',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours: ')

std_hr = [[float(study_hrs)]]


ans = model.predict(std_hr)
marks = round(ans[0],2)
if marks > 100:
    marks = 100 
elif marks<0:
    marks = 0 
else:
    marks = marks
print(f'Your Predicted marks is: {ans[0]}')
```
