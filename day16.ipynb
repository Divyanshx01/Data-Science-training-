```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("gurdit559/canada-per-capita-income-single-variable-data-set")

print("Path to dataset files:", path)
```

```python
#Step 3:  File Read
file_path = path + '/' + os.listdir(path)[0]

#Step 4:  Read data
df = pd.read_csv(file_path)
print('Done')
```

```python
X = df[['year']]
y = df['per capita income (US$)']
print('Done')
```

```python
df[['year']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
print('Model Created')
```

```python
model = LinearRegression()

#training Step
model.fit(X,y)  # Hyper Parameters: Model Fine Tune
```

```python
year = [[2018]]
capita = round(model.predict(year)[0],2)
print(f''' per capita $
{year[0][0]}:{capita}''')

```

```python
y_pred = model.predict(X)
print('Done')
```

```python
df.sample()
```

```python
df.head(10)
```

```python
print(df)
```

```python

df['Predicted Capita'] = y_pred
print('Done')
```

```python
(df['per capita income (US$)'] - df['Predicted Capita']).abs().mean()
```

```python
from sklearn.metrics import mean_absolute_error
print('Done')
```

```python
#Step:  Evalute()

# mean_absolute_error(y_actual, y_pred)
# y_actual = y
# y_pred = y_pred


MAE = mean_absolute_error(y, y_pred)
```

```python
print('Mannual mae:',MAE)
print('Actual mae:',MAE)
```

```python
(df['per capita income (US$)'] - df['Predicted Capita']).abs().mean()
```

```python
mse = ((df['per capita income (US$)'] - df['Predicted Capita'])**2).mean()
mse
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE = mean_squared_error(y,y_pred)
print(MSE)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
rmse = mse**.5
RMSE = root_mean_squared_error(y,y_pred)
```

```python
print('Mannual mse:',rmse)
print('Actual MSE:',RMSE)
```

```python
model_score = model.score(X,y)
print('Model score is:',model_score)
```
