```python
class Car:

    def __init__(self):
        self.color = 'Black'
        self.wheel = 4
        self.spare_wheel = 1
        self.airbag_support = True
        self.fuel_type = 'Petrol'
        self.car_type = 'Sedan'
        self.gear_type = 'Automatic'
        print('Your car is ready')

    # ====================================
    #Features

    def auto_park(self):
        import IPython
        url = '''https://media.tenor.com/fpwy0T0eW8cAAAAM/classic-mini.gif'''
        display(IPython.display.Image(url))

    def engine_on(self):
        import pygame
        pygame.init()
        path = r'''/Users/divyanshchoudhary/Downloads/chevrolet-corvette-c6-sound-effect-360531.mp3'''
        music = pygame.mixer.Sound(path)
        music.play()

    def speed_up(self):
        c = 10
        while c<=200:
            import time
            print(f'Speed is Increasing by {c} km/hr')
            c+= 10
            display(clear= True)
            time.sleep(0.5)

            if c >=100:
                print('You are crossing speed limit')
                break

    def drift_brake(self,drift = False):
        if drift == True:
            import pygame
            pygame.init()
            path = r'''/Users/divyanshchoudhary/Downloads/1JZ ALTEZZA drifting sounds.mp3'''
            music = pygame.mixer.Sound(path)
            music.play()
            import IPython
            url = '''https://media.tenor.com/puPGAEN97U0AAAAM/lol-mekro.gif'''
            display(IPython.display.Image(url))

        else:
            print('Brake Applied!!')
```

```python
class ATM:

    def __init__(self):
        print("Object created successfully!")
        self._pin = 1111 # protected variable 
        self._balance = 2000 # protected variable

    def change_Pin(self): # protected method
        self.pin = int(input("Enter new PIN:"))
        print("Your PIN has been changed")

    def show_Balance(self):
        print(f'''Your Current Balance is {self._balance}''')
        
    def _atm_Performance_Status(self):
        print("ATM working fine 100%")
    
```
