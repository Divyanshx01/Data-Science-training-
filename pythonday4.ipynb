```python
from googlesearch import search

title = 'Chat GPT Trending News and Investments'

data = search('title',num_results=7)

all_data = []

for i in data:
    all_data.append(i+'\n')

    # Mode = a:It will append new line

    
    file = open (f'{title}.txt',mode = 'w')
    file.write('===============================\n')
    file.writelines(all_data)
    file.close()
    print('Done')
```
